Graph of a function
-------------------

Graph of a function $f:\mathbb R \to \mathbb R$ is a set of all pairs of the form $(x,f(x))$. We will use vector wise functions of NDarray to generate lots of points on the graph. But first let's enable IPython's magic command *%pylab*

```python
%pylab inline
```

Let's draw a graph of the parabola given by the quadratic function $f(x)=1-x^2$ on the interval $[-1,1]$.

```python
f = lambda x: polyval(array([-1, 0, 1]),x) # polynomial can be given by a vector of the coefficients
x = linspace(-1,1,100) # a vector with 100 equidistant points between -1 and 1
y = f(x)
graph = plot(x,y)
```

#### Next we have to calculate the tangent to a point. Let's choose a point with $x_0=\frac{1}{2}$. The tangent line can be described by a linear function  $$y=k_tx+n,$$  where $k_t=f'(x_0)$ and the free term $n$ is determined so that the point $(x_0,f(x_0))$ lies on the tangent line.

```python
x0 = 0.5
y0 = f(x0)
df = lambda x: polyval(array([-2, 0]),x)
kt = df(x0)
n = y0 - kt*x0
plot(x,y,x,kt*x+n)
```
